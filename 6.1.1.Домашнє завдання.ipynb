1. Ви —  продакт менеджер в компанії, яка займається мобільними додатками. Ваша модель монетизації — підписка. У травні 2019 року до вас приєдналося 100 підписників, у червні — 150, у липні — 170. Скільки ці користувачі принесли вам до початку травня 2021 року (за два роки), якщо щомісячний рівень відтоку становить 5%, а підписка коштує $10?
   
Для визначення доходу принесеного до початку травня 2021 року користувачами, які зареєструвались в травні-липні 2019, розрахумо загальний дохід по кожному місяцю реєстрації:
- дохід від користувачів зареєстрованих в травні 2019 року  до початку травня 2021 року (24 місяці):
      травень 2019 - 100 підписників * $10 = $1000 
      червень 2019 - квітень 2021:  10 * 100 * ((1 - 0.95**23) / (1-0.95)) = $13852.86
      Всього: $14852,86
- дохід від користувачів зареєстрованих в червні 2019 року  до початку травня 2021 року (23 місяці):
      червень 2019 - 150 підписників * $10 = $1500 
      липень 2019 - квітень 2021:  10 * 150 * ((1 - 0.95**22) / (1-0.95)) = $ 20293,99
      Всього: $21793,98
- дохід від користувачів зареєстрованих в липні 2019 року  до початку травня 2021 року (22 місяці):
      липень 2019 - 170 підписників * $10 = $1700 
      серпень 2019 - квітень 2021:  10 * 170 * ((1 - 0.95**21) / (1-0.95)) = $22420.90
      Всього: $24120.90
Отже, отриманий дохід  за період травень 2019 - початок травня 2024 від зареєстрованих користувачів в травні, червні, липні 2019 року становить $60767.74.


2.  Ви — менеджер по роботі з продуктом в компанії, яка розробляє мобільні ігри. У січні 2020 року у вас було 2000 активних користувачів. Щомісяця кількість нових користувачів зростає на 3%, але водночас 2% існуючих користувачів перестають користуватися додатком. У результаті, ефективне зростання активних користувачів складає 1% щомісяця.
    
    Ваша мета — визначити кумулятивну кількість активних користувачів за весь період від січня 2020 року до кінця грудня 2021 року (тобто за 24 місяці), враховуючи щомісячний приріст 1%.
    
    **Питання**: Якою буде загальна кількість активних користувачів, накопичена за 24 місяці?

     Маючи початкову кількість підписників в січні 2020 року у розмірі 2000 активних коритувачів і відомий щомісячний приріст у розмірі 1% визначимо кумулятивну кількість активних користувачів за період  січень 2020 року - грудень 2021 року (включно):
    2000 * (1 - 1.01**24) / (1-1.01) = 53947
    Отже, за період  січень 2020 року - грудень 2021 року кумулятивна кількість активних користувачів становила 53947 коритувачів.


3. Ви — менеджер з розвитку бізнесу в компанії, що займається створенням платформи для онлайн-курсів. У січні 2021 року ви залучили 500 нових користувачів. У лютому кількість нових користувачів зросла на 50%, у березні — ще на 20%. З квітня приріст нових користувачів стабілізувався на рівні 10% щомісяця. Скільки активних користувачів ви мали загалом за перше півріччя 2021 року? 

Визначимо кількість активних користувачів для кожного пріоду окремо:
- січень 2021 року - 500 користувачів;
- лютий 2021 року - враховучи ріст користувачів на 50 % відносно січня 2021 року, то їх кільсть становила - 500 * 1,5 = 750 користувача;
- березень 2021 року - враховучи ріст користувачів на 20 % відносно лютого 2021 року, то їх кільсть становила - 750 * 1,2 = 900 користувачів;
- квітень - червень 2021 року (включно) - враховучи щомісячний приріст користувачів на 10 % відносно березня 2021 року, то їх кільсть становила  
   900 * (1 - 1.1**3) / (1-1.1) = 2979 підписників
  Підрахувавши суму значень кількості активних користувачів для визначеного для кожного періоду окремо визначимо загальне значення за перше півріччя 2021 року активних користувачів  у кількості 5129 користувачів.
